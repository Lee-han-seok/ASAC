# 4. 흐름제어

- 목표
  - 프로그램의 흐름을 제어하여 어떤 목적을 달성하는 것

- 연산자
  - 연산자 표현, 의미, 우선순위
  - +, -, *, /, <, >, ==, ...

- 조건문
  - 조건문에 의해서 처리 => 결론 참 아니면 거짓
  - 거짓상황
    - None, , 0.0, Flase, [], (), {}, 기타연산(비교) 결과 -> 코드가 간결해짐 (Python은 최적화된 코드를 짤 필요가 있음, C나 C++처럼 짤 필요 없음)
  - 같이 활용되는 예약어/키워드(변수로 사용 X)
    - if, else, elif, ...

- 반복문
  - 반복적인 업무를 수행 -> 구구단
  - 한정된 반복 -> 번만 반복해
    - for
  - 무한 반복 -> 언제 끝날지 모르겠지만
    - while
  - break : 나를 감싸고 있는 가장 가까운 반복문을 탈출한다. 통상적으로 조건문과 같이 사용
  - continue : 통상적으로 조건문과 같이 사용, 가장 가까운 반복문으로 점프

- 조합 => 흐름제어
  - 조건식 + (if ~ for or while ~ break or continue) 등을 중복적으로 필요에 의해 조합하여 원하는 흐름을 생성, 관리
