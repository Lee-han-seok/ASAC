# 조건문, 조건식

- 조건식
  - 조건식의 결론 => True or False
  - False 상황 : 0, ....

- 조건식에 사용하는 연산자
  - 값 비교
    - A == B : A와 B는 일치(동일, 같은)하는가?
    - A != B : A와 B는 다른가?
    - A > B  : A가 B보다 값이 큰가?
    - A < B  : A가 B보다 값이 작은가?
    - A >= B : A가 B보다 값이 크거나 같은가?
    - A <= B : A가 B보다 값이 작거나 같은가?
    - not A : A값을 부정하다

  - 조건식 끼리 비교
    - (A == B) 연산자 (A > B)
    - C1 or C2 : 하나만 참이면 끝 -> 참임
      - C1이 참이면 C2는 체크 않함
      - 보다 빨리 연산처리가 됨
    - C1 and C2 : 하나만 거짓이면 끝 -> 거짓
      - C1이 거짓이면 C2체크 않함
      - 둘다 참이여야만 참이다
    - not C1 : 부정을 잡는 조건식
      - False 상황을 체크하고 싶다면
      - not False => True

  - 기타
    - A is B : 객체의 id값 비교
    - a in A : A안에 a가 있는가? (맴버인가요?)

````python
# 값에 대한 조건식
a = 10
b = 11

# 조건식의 결론은 참과 거짓이다 => 출력값도 참과 거짓이다
a == b, a != b, a > b, a < b, a >= b, a <= b
````

````python
# 조건식 대 조건식

# a가 b보다 작거나 혹은 a가 b와 같다면 -> 참이다
# ~ 거나 ~  => or
a < b or a == b  # True or a == b => 참 확인 => True 리턴
````

````python
# a와 b는 같고, 그리고 a는 b보다 작다
# ~ 고 ~ => and
a == b and a < b # False and a < b => False 발견 => False 리턴
````

#### range()

- 수치를 제너레이터 방식으로 공급하는 함수

````python
# 0 <= x < 3
# range(3) == range(0, 3), 시작값은 생략가능(생략하면 0부터)
for i in range(3):
    print( i, end=',' )

# 1 <= x < 3
for i in range(1, 3):
    print( i, end=',' )

# 1 <= x < 10, step=2
for i in range(1, 10, 2):
    print( i, end=',' )
````

## 컴프리핸션(내포)



- 리스트 내포, 딕셔너리 내포
- 결과를 중심으로 간결하게 구성
- 결과물은 정적구성 방식 활용( [ .. ], { .. } )
- 파이썬 2.0부터 지원, 딕셔너리 2.7부터 지원

#### 리스트 컴프리핸션

````python
# 아래 코드 상에서
# 프린트로 출력한 결과를 -> 담은 내용
# 리스트에 담아서 -> 리스트에 추가(append)
# 출력(상위값 3개만) -> 리스트의 슬라이싱 수행 -> 리스트
tmps = list()

for n in range(2,10):
    for m in range(1,10):
        tmps.append( f'{n:2} x {m:2} = {n*m:>2}' )

tmps[ : 3]
````

````python
# 1. [] 준비
# 2. [] 내부에 담길내용(데이터) or 형태(구조, 포멧팅) 채운다
# 3. 담길내용에 사용된 변수들을 정의한 구조를 가져온다 (for문)

[ f'{x:2} x {y:2} = {x*y:>2}' for x in range(2,10) for y in range(1,10) ][:3]
````

````python
# 반복문에 조건식 추가된 유형
# 4단 7단 제외
# 조건식 n이 4거나 7이면 다시 반복문으로 점프
tmps = list()

for n in range(2,10):
    if (n==4) or (n==7) :continue
    for m in range(1,10):
        tmps.append( f'{n:2} x {m:2} = {n*m:>2}' )

tmps
````

````python
# continue 미사용으로 구현
tmps = list()

for n in range(2,10):
    if (n==4) or (n==7):
        pass
    else:
        for m in range(1,3):
            tmps.append( f'{n:2} x {m:2} = {n*m:>2}' )

tmps
````

````python
# continue 미사용으로 구현
tmps = list()

for n in range(2,10):
    if (n!=4) and (n!=7):
        for m in range(1,3):
            tmps.append( f'{n:2} x {m:2} = {n*m:>2}' )

tmps
````

````python
# 아래 표현으로는 에러가 발생함
[ f'{x:2} x {y:2} = {x*y:>2}' for x in range(2,10) if (x!=4) and (x!=7) for y in range(1,3) ]
````

## break, continue, else



- break :
    - 가장 가까운 반복문을 탈출한다
    - 특정 조건문을 같이 사용하여 특정 조건시에만 탈출하게 사용 가능
    ```
        if 조건문:
            break
    ```    
- continue :
    - 가장 가까운 반복문으로 점프한다
    - 특정 조건문을 같이 사용하여 특정 조건시에만 점프하게끔 사용 가능
    ```
        if 조건문:
            continue
    ```
- else
    - for, while, try 에서 같이 혼용 사용 가능
    - 의미
        - 반복문 : 정상적으로 모두 순환해서 수행했다것을 인증, 확인하는 코드 -> 이 인증값(순환이 모두 수행되었다)이 필요하다면
            -> 체크 포인트
        - 예외처리 : 오류가 없었다라는 의미


````python
# break 사용
for n in range(3,6):# 3+3):
    for m in range(1,10):
        print( f'{n:2} x {m:2} = {n*m:>2}' )
        break # 반복문안에서 1회만 작동하게 하는 스타일
````

````python
# 4단 제외, 3,5단만 출력
for n in range(3,6):# 3+3):
    if n == 4:
        continue
    for m in range(1,10):
        print( f'{n:2} x {m:2} = {n*m:>2}' )
````

````python
# m값이 5이상이면 반복 중단 처리하시오(1,2,3,4만 출력)
%%time
for n in range(3,6):# 3+3):
    for m in range(1,10):
        if m >= 5:break
        print( f'{n:2} x {m:2} = {n*m:>2}', end=' ' )
````

## while문

- 무한루프를 염두하고 사용되는 코드
    - 0 ~ 무한대 반복
    - 반드시 탈출하는 코드가 필요하다

- 문법
    ```
        while 조건식:
            statements
        
        ex1)
        while True:
            statements
            ...
            if 조건식:
                break
        
        ex2)
        while 조건식:
            statements
            if 조건식:
                break or 변수값 조작

    ```

````python
# 기본형
datas = [1,2,3]
while len(datas) > 0 : #  데이터가 하나라도 있으면 뭔가를 수행해라
    print( 'hi' )
    break
````

````python
# 데이터가 하나라도 있으면(변수를 그냥 조건식에 넣어라) 뭔가를 수행해라 -> 조건식 변경
while datas:
    print( 'hi' )
    break
````
