# 3. 딕셔너리

- 순서X, Key & Value의 쌍으로 관리된다.
- 텍스트 데이터 포멧(xml, json...)중 json과 바로 대응이 된다.
  - db -> 쿼리 -> 딕셔너리 -> json -> 전송 -> 수신 -> javascript 객체로 받는다.(상호호환이 가능하다)

```python
# 생성
dic = dict() # 나중에 맴버를 추가할 때 조금 더 어룰린다.

dic, type(dic), len(dic)
```

```python
# 생성
dic = {
    # 키 : 값 형태
    # 키 -> 주로 단일 데이터형 사용
    # 값 -> 모든 타입이 올 수 있다.
    'name' : '공덕역'
} # 맴버를 추가하면서(세팅) 딕셔너리를 바로 생성할 때 좀 더 어울린다.
dic, type(dic), len(dic)
```

```python
# 맴버추가 및 수정도 가능하다.
dic = dict()
dic['키'] = '값'
dic
```

```python
# 생성
dic = {
    'name' : '공덕역'
}
dic, type(dic), len(dic)
```

```python
# 인덱싱 -> 값을 추출
# 변수 [키]
dic[ 'name' ]
```

```python
dic[0] # 사전형은 순서가 없고 인덱스도 없다. 즉 입력되는 값은 key값을 호출한것이고  0이 없기 때문에 오류 발생
```

```python
dic[0] = 'hi' # 딕셔너리에 값 추가
# 위에서 0번키를 생성하고 값을 세팅했으므로, 0번키를 넣어서 인덱싱(추출) 가능
dic, dic[0]
```

```python
# 슬라이싱 -> X(순서가 없음 -> 범위 설정 못함)
```

```python
# 추출
dic.get('name')
```

```python
dic.get(0)
```

```python
# 해당키가 없으면 None 출력
print(dic.get('age'))
```

```python
# 키가 없으면 셧다운
dic['age'] # 에러 발생
```

# 4. 튜플


- 리스트와 동일, 단 수정 불가
- 용도는 통상적으로 함수가 여러개의 값을 리턴한다면 그 타입이 튜플이 된다.
- 함수의 가변인자도 튜플타입니다.
- 직접 만드는 경우보다는 자동으로 생겨서 사용되는 경우가 많다.
- 값을 묶는다
- 주의사항
  - 맴버가 1개이면 => (1,) : 이렇게 반드시 ','추가
  - (1)은 값 1을 의미

```python
# 생성
tu = tuple() # 의미가 없음 : tuple은 수정 안되니까!
tu, type(tu), len(tu)
```

```python
# 데이터를 넣으면서 생성하는 경우는 사용가능
tu = (1,2,3)
tu, type(tu), len(tu)
```

```python
# 인덱싱
tu[ 0 ]
```

```python
# 슬라이싱
tu[:2] # 순서 있으니까 실행 가능
```

```python
# 주 용도
# tu의 값들 중 2번째것만 받고 나머지는 버려라

_, v, _ = tu
v
```

```python
# 튜플은 수정 불가야!!
tu[0] = 100
```

# 5. 집합

- 중복 제거, unique

```python
# 중복된 데이터 준비

a = list('hellowworld')
a
```

```python
# list를 데이터 넣어서 set을 생성
# 중복 제거 및 순서 변경
set( a )
```

```python
# 2개의 데이터를 준비해서 교집합, 차집합, 합집합
a1 = [1,2,5,6,7,3,7,8,9,2,4,22,77]
b1 = [1,3,5,7,9,1,7,7,6,5,4]
```

```python
a = set(a1)
b = set(b1)

a,b
```

```python
# 교집합
a.intersection(b)
```

```python
# 합집합
a.union(b)
```

```python
# 차집합(빼는 방향 중요)
a.difference(b), b.difference(a)
```
