### 3-3-5. 포멧팅

- 문자열과 문자열을 결합할 때
- 문자열 내부에 특정 데이터를 삽입할 때
- 특정 포맷을 따르게 문자열을 구성할 때
- 방식(3가지)
  - '문자열' % (값, 값, ...)
  - '문자열'.format(값, 값, ...)
  - f'문자열...{삽입문자열}...{삽입문자열}..'
    - fstring : python 3.6부터 추가
    - 구조를 유지하면서 포멧팅 지원


```python
# 포멧팅 1번 유형 => %
a = 1
b = 2
# 출력하시오 : '1 + 2 = 3' 으로 출력
# 문자열 내부에 값을 넣을 때는 %d : 정수, %f : 부동소수, %s : 문자열, 모든 타입을 받을 수 있다.
'%d + %d = %d' % (a, b, a+b)
'%s + %s = %s' % (a, b, a+b) # 타입을 모르면 %s로 받아도 된다.
```

```python
# 포멧팅 2번 유형 처리
# 포멧팅 2번 유형 처리
'{0} + {1} = {2}'.format(a,b,a+b)
```

```python
# 포멧팅 2번 유형 처리
'{1} + {0} = {2}'.format(a,b,a+b) # 순서 조작 가능
```

```python
# 인자값들의 의미가 애매모호하다 -> 인자에 이름을 부여(파라미터명)해서 명확하게 사용할 수 있다.

'{x} + {y} = {result}'.format(x = a, y = b, result = a+b)
```

```python
# 포메팅 3번으로 적용
f'{a} + {b} = {a+b}'
```

```python
# 각 값 자리에 크기, 정렬(왼쪽, 오른쪽, 중앙) 지정, 공백 대신 기호로 채우기도 가능하다
f'{a:10} + {b:*<10} = {a+b:*^11}'
```
