## 3-1. 기본다루기

```python
# 한줄 문자열
a = 'abcd ABCD 1234 !@#$ 가나다라'
a
# 코랩에서 문제없이 잘 출력됨
```

```python
# 타입 확인
type(a)
```

```python
# 여러줄 구조 유지 가능
a = '''
  select
    *
  from
    users:
'''

print(a)
```

```python
# 섞어서 표현 -> 문자열 기호도 문자로 인식
a = 'abc\"de\"fg' # \을 사용하면 내부의 문자열 기호와 외부의 기호가 동일해도 사용할 수 있음. (\ : 이스케이프 문자)
print(a)
```

```python
# 문자열간 + 연산 (더하기)
a = 'abc'
b = 'def'

a+b, a, b
```

```python
# 문자열 반복
a = '-'
print(a * 3) # 곱하기로 반복 가능
```

## 3-2. 인덱싱

- 문자열에서 특정 문자열을 **추출**하는 행위
- 연속형 데이터 타입에는 모두 적용가능
- 데이터 -> 인덱싱 -> 차원축소가 진행
  - 2D -> 1D
- 문자열은 차원축소 X

```python
a = '0123456789'
a
```

```python
# 요구사항 - a라는 문자열이 가진 값들 중에서 값이 1인 값을 추출하시오
# 문법 : 변수명[ 인덱스 ] -> 가까운 쪽에서 접근합시다
# 정방향 인덱스값을 이용하여 추출하시오
a[1] # 순서에 맞게
```

```python
# 역방향은?
a[-9]
```

## 3-3. 슬라이싱

- 원본 데이터에서 특정 데이터 부분을 자른다
- 잘라서 추출한 데이터는 원본과 차원이 동일
  - 차원유지 : 2d -> 슬라이싱 -> 2d
- 문법
  - 변수명[ 시작인덱스 : 끝인덱스 : 간격(step, 기본값은 1 생략 가능)  ]

```python
a
```

```python
a[ : ] # 원본 카피
```

```python
a[:2]
```

```python
# 뒤에서부터 2개 데이터를 슬라이싱
a[-2:]
```

```python
# 앞에서 1개 뒤에서 1개
a[1:-1]
```

```python
#앞에서 1개 뒤에서 1개 제외한 데이터에서 하나씩 건너서 슬라이싱
a[1:-1:2]
```
