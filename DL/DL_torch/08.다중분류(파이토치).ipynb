```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential,CrossEntropyLoss, Softmax
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/softmax.txt', sep = ' ', header = None)
df
```

```python
x_data = df.iloc[:,0:2]
y_data = df.iloc[:,2:]
```

```python
x = torch.FloatTensor(x_data.values)
y = torch.FloatTensor(y_data.values)
```

```python
model = Sequential()
model.add_module('nn1', Linear(2,3))
model.add_module('soft1', Softmax(dim = 1))
loss_fn = CrossEntropyLoss()
optimizer = Adam(model.parameters(), lr = 0.01)
```

```python
for step in range(2000):
    optimizer.zero_grad()
    hx = model(x) # z = matmul(x, w) + b => hx = softmax(z)
    cost = loss_fn(hx, y)
    cost.backward()
    optimizer.step()
    print(step, cost.item())
```
