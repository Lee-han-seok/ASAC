```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
print(2**2)
print(2**-2)
print(2**3)
print(2**-3)
print(math.e)
```

```python
def sigmoid(z) :
    return 1 / (1 + math.e ** (-z))
```

```python
x = np.arange(-10,10,0.1)
p = 1 / (1 + np.exp(-x))

plt.plot(x, p, 'r--')
plt.xlabel('x')
plt.ylabel('1 / (1 + e^-x)')

plt.show()
```

```python
for z in np.linspace(-10,10,50):
    s = sigmoid(z)
    print('z=', z, 's=', s)
    plt.plot(z, s, 'go')
plt.show()
```
