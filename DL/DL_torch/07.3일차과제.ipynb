```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from scikeras.wrappers import KerasRegressor
```

# 1. 일반 정규화 (pipeline X)

```python
train = pd.read_excel('data/아버지아들키.xlsx', sheet_name = 'train')
test = pd.read_excel('data/아버지아들키.xlsx', sheet_name = 'test')
```

```python
train
```

```python
X_train = train.iloc[:,[0]].values
y_train = train.iloc[:,[1]].values
X_test = test.iloc[:,[0]].values
y_test = test.iloc[:,[1]].values
```

```python
train.corr()
```

```python
scaler_X = MinMaxScaler()
X_train_scaled = scaler_X.fit_transform(X_train)

scaler_Y = MinMaxScaler()
y_train_scaled = scaler_Y.fit_transform(y_train)
```

```python
X_test_scaled = scaler_X.transform(X_test)
y_test_scaled = scaler_Y.transform(y_test)
```

```python
# 정규화
dense = Dense(units = 1, input_dim = 1)
model = Sequential([dense])
model.compile(loss = 'mse', optimizer = Adam(0.01))
h = model.fit(X_train_scaled, y_train_scaled, epochs = 1000)
```

```python
model.predict(X_test_scaled)
```

```python
y_pred=scaler_Y.inverse_transform(model.predict(X_test_scaled))
```

```python
r2_score(y_test, y_pred)
```

# 2. pipeline

```python
X_train = train.iloc[:,[0]].values
y_train = train.iloc[:,[1]].values
X_test = test.iloc[:,[0]].values
y_test = test.iloc[:,[1]].values
```

```python
def createModel():
    dense = Dense( units=1, input_dim=1)
    model = Sequential( [dense] )
    model.compile( loss='mse', optimizer=Adam( 0.1) )
    return model
```

```python
kcf = KerasRegressor(model=createModel , epochs=1000, verbose=1 )
```

```python
pmodel = make_pipeline( StandardScaler(), kcf)
pmodel.fit( X_train, y_train )
```

```python
pred_train = pmodel.predict( X_train)
pred_test = pmodel.predict(X_test)
```

```python
print(r2_score(y_train, pred_train))
print(r2_score(y_test, pred_test))
```
