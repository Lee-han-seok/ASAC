아버지와 아들키

- 아버지 키가 170인경우 아들키 예측
- train, test
- 정규화
- (케라스 버전)
- 파이프라인
- 그냥 (파이토치 하이레벨) 

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scikeras.wrappers import KerasRegressor
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
df_train = pd.read_excel("아버지아들키.xlsx",sheet_name=0)
df_test = pd.read_excel("아버지아들키.xlsx",sheet_name=1)
```

```python
df_train.head()
```

```python
df_train.shape, df_test.shape
```

```python
X_train = df_train[['Father']]
X_test = df_test[['Father']]
y_train = df_train[['Son']]
y_test = df_test[['Son']]
```

# 파이프라인

```python
def createModel():
    dense = Dense( units=1, input_dim=1)
    model = Sequential( [dense] )
    model.compile( loss='mse', optimizer=Adam( 0.1) )
    return model
```

```python
kcf = KerasRegressor(model=createModel , epochs=1000, verbose=1 )
```

```python
pmodel = make_pipeline( StandardScaler(), kcf)
pmodel.fit( X_train, y_train )
```

```python
pmodel.predict([[170.0]])
```

```python
pred_train = pmodel.predict( X_train)
pred_test = pmodel.predict(X_test)
```

```python
print(r2_score(y_train, pred_train))
print(r2_score(y_test, pred_test))
```

# 케라스 그냥

```python
# 정규화
X = pd.concat([X_train, X_test])
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train = X[:755]
X_test = X[755:]
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
dense = Dense(units = 1, input_dim = 1)

model_ = Sequential([dense])
model_.compile(loss = "mse", optimizer = Adam(learning_rate = 0.01))
model_.fit(X_train, y_train, epochs = 1000)
```

```python
model_.predict(scaler.transform([[170.0]]))
```

```python
pred_train_ = model_.predict(X_train)
pred_test_ = model_.predict(X_test)

print(r2_score(y_train, pred_train_))
print(r2_score(y_test, pred_test_))
```

# 파이토치

```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_excel("아버지아들키.xlsx",sheet_name=0)
df_test = pd.read_excel("아버지아들키.xlsx",sheet_name=1)
```

```python
X_train = df_train[['Father']]
X_test = df_test[['Father']]
y_train = df_train[['Son']]
y_test = df_test[['Son']]
```

```python
X = pd.concat([X_train, X_test])
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train = X[:755]
X_test = X[755:]
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train.to_numpy()
```

```python
X_train = torch.FloatTensor(X_train)
```

```python
linear = Linear(in_features = 1, out_features =  1) # w : 1x1 , b : 1 , hx = w*x +b
model_py = Sequential()
model_py.add_module('nn1', linear)
loss_fn = MSELoss()
optimizer = Adam(model_py.parameters(), lr = 0.1)

for step in range(1000):
    optimizer.zero_grad()
    hx = model_py(X_train) # w*x + b / linear(x)
    cost = loss_fn(hx, y_train.to_numpy()) # 이게달라짐
    cost.backward()
    optimizer.step()
    print(step, cost.item())
```
