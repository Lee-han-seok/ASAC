```python
import torch
import matplotlib.pyplot as plt
```

```python
x_data = torch.FloatTensor([1,2,3])
y_data = torch.FloatTensor([1,2,3])
```

```python
def cost(x,y,w):
    hx = w*x # [-1,-1,-1 ] X [1,2,3]
    s = torch.sum((hx-y)**2)
    return s/len(x)
```

```python
def gradient(x,y,w):
    hx = w*x
    s = torch.sum((w*x-y)*2*x)    
    return s/len(x)
```

```python
cost(x_data, y_data, -1)
```

```python
w = torch.tensor(10)
for epoch in range(200):
    c = cost(x_data, y_data, w)
    print('epoch:', epoch, 'cost:', c.item(), 'w=', w.item())
    g = gradient(x_data, y_data, w)
    w = w - 0.1*g
print('최종:', w)
```
