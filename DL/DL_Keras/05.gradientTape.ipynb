```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
x = tf.Variable(4.0)
with tf.GradientTape() as tape:
    y = x**2

dy_dx = tape.gradient(y,x)
dy_dx
```

```python
x = tf.constant([1,2,3,4,5], tf.float32)
y = tf.constant([3,5,7,9,11], tf.float32)
```

```python
w = tf.Variable(10.0, dtype = tf.float32)
b = tf.Variable(10.0, dtype = tf.float32)
```

```python
for i in range(2000):
    with tf.GradientTape() as tape:
        hx = w*x + b
        cost = tf.reduce_mean((hx-y)**2)
    w_grade, b_grade = tape.gradient(cost, [w,b])
    w.assign_sub(0.01*w_grade)
    b.assign_sub(0.01*b_grade)
    print('cost:',cost)
print('w:', w.numpy())
print('b:', b.numpy())
```
