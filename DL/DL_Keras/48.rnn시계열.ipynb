```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU 
# SimpleRNN -> LSTM (계산이 너무 복잡) -> GRU (계산 간소화, LSTM 정도의 성능) 
from tensorflow.keras import Sequential

import pandas as pd
```

```python
df = pd.read_csv('./data/sample2.csv', index_col = 'date')
```

```python
def make_dataset(data, label, window_size=20):
    feature_list = []
    label_list = []
    for i in range(len(data) - window_size):
        feature_list.append(np.array(data.iloc[i:i+window_size]))
        label_list.append(np.array(label.iloc[i+window_size]))
    return np.array(feature_list), np.array(label_list)
```

```python
df
```

```python
x, y = make_dataset(df[['a', 'b']], df[['a', 'b']], 3)
```

```python
x.shape
```

```python
y.shape
```

```python
model = Sequential()
model.add(SimpleRNN(16, activation = 'relu', input_shape = (3,2)))
model.add(Dense(units = 32))
model.add(Dense(units = 16))
model.add(Dense(units = 2))
model.compile(loss = 'mse', optimizer = 'adam')
model.fit(x, y, epochs = 500)
```

```python
xdata = np.array([[[1,21],[2,22],[3,23]]])
xdata.shape
```
