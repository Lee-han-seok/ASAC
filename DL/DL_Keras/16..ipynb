```python
df = pd.read_csv('data/test.csv', header = None)
df.columns = ['q1','q2','mid','final']
```

```python
x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,-1].values
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)

w = tf.Variable(tf.random.uniform([3,1]))
b = tf.Variable(tf.random.uniform([1]))
```

```python
def compute_loss():
    hx = tf.matmul(x, w) + b
    cost = tf.reduce_mean((hx-y)**2)
    return cost
```

```python
patience = 50
min_delta = 0.01
hist_loss = []
pcnt = 0

optimizer = Adam(0.1)
for i in range(10000):
    optimizer.minimize(compute_loss, var_list = [w,b])
    c = compute_loss().numpy()
    hist_loss.append(c)
    print(i, 'cost:', c)
    if i > 0:
        if hist_loss[i-1] - hist_loss[i] > min_delta:
            pcnt = 0
        else :
            pcnt += 1
        if pcnt >= patience :
            print('early stopping')
            break

# 지정한 min_delta보다 앞뒤의 차이가 크면 0을 더하고, 그렇지 않다면 1을 더함(pcnt)
# loss값이 min_delta값보다 적을 경우 학습이 다 되었다고 판단하는것임!
# 쭉 학습을 진행하며 쌓인 값이 미리 지정한 임계치(patience)
# min delta는 0.1 혹은 0.001
```

```python
dense = Dense(units = 1, input_dim = 3)
model = Sequential([dense])
model.compile(loss = 'mse', optimizer = Adam(0.1))
earlyStop = EarlyStopping(monitor = 'loss', min_delta = 0.01, patience = 100)
h = model.fit(x_data, y_data, epochs = 1000, callbacks = [earlyStop])
```
