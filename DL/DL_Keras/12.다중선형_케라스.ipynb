```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('data/test.csv', header = None)
df.columns = ['q1', 'q2', 'mid', 'final']
df
```

```python
x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,-1].values
```

```python
x_data.shape
```

```python
dense = Dense(units = 1, input_dim = 3) # w : 3 X 1, b : 1
```

```python
dense(x_data) # hx = matmul(x_data, w) + b
```

```python
w, b = dense.get_weights()
w, b
```

```python
dense = Dense(units = 1, input_dim = 3) # w : 3 X 1, b : 1
model = Sequential([dense])
model.compile(loss = 'mse', optimizer = Adam(0.1))
model.fit(x_data, y_data, epochs = 500)
```

```python
#dense(x_data)
model.predict(x_data)
```

```python
from sklearn.metrics import r2_score

r2_score(y_data, model.predict(x_data))
```
