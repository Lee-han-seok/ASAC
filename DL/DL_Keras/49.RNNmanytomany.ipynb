```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU, TimeDistributed
# SimpleRNN -> LSTM (계산이 너무 복잡) -> GRU (계산 간소화, LSTM 정도의 성능) 
from tensorflow.keras import Sequential

import pandas as pd
```

# 1. Many to One

```python
x = np.array([[1,2,3],[2,3,4],[3,4,5],[4,5,6]])
y = np.array([4,5,6,7])
```

```python
xR = x.reshape(4,3,1) # batch, timestep, 특성데이터의 수
xR
```

```python
model = Sequential()
model.add(SimpleRNN(16, activation = 'tanh', input_shape = (3,1)))

model.summary()
```

```python
model = Sequential()
model.add(LSTM(16, activation = 'tanh', input_shape = (3,1)))
```

```python
model.summary()
```

```python
model = Sequential()
model.add(GRU(16, activation = 'relu', input_shape = (3,1)))

model.summary()
```

```python
model = Sequential()
model.add(GRU(16, activation = 'relu', input_shape = (3,1)))
model.add(Dense(units = 32))
model.add(Dense(units = 16))
model.add(Dense(units = 1))
model.compile(loss = 'mse', optimizer = 'adam')
model.fit(xR, y, epochs = 500)

model.summary()
```

```python
 model.predict(np.array([5,6,7]).reshape(1,3,1))
```

# 2. Many to Many

```python
X = np.array([x for x in range(5, 301, 5)]).reshape(20,3,1)
Y = np.array([y for y in range(20, 316,5)]).reshape(20,3,1)
```

```python
X
```

```python
Y
```

```python
model = Sequential()
model.add(LSTM(64, activation = 'tanh', input_shape = (3,1), return_sequences = True)) # return_sequences = True
model.add(Dense(32))
model.add(Dense(16))
model.add(TimeDistributed(Dense(1))) # timestep별로 정답을 매칭 시켜줌 ㅇㅇ
model.compile(loss = 'mse', optimizer = 'adam')
```

```python
model.summary()
```

```python
model.fit(X,Y, epochs = 1000)
```

```python
model.predict(np.array([300,305,310]).reshape(1,3,1))
```
