```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.array([1,2,3,4,5,6])
y_data = np.array([0,0,0,1,1,1])
```

```python
def cost(x, y, w):
    z = w * x
    hx = tf.nn.sigmoid(z) # 0 ~ 1

    cost_i = -tf.reduce_mean(y*tf.math.log(hx) + (1-y)*(tf.math.log(1-hx)) )

    c = tf.reduce_mean(cost_i)
    return c
```

```python
for w in np.linspace(-3,5, 50):
    c = cost(x_data, y_data, w)
    plt.plot(w, c, 'ro')
plt.show()
```
