```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential, Model
from tensorflow.keras.layers import Dense, Input
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0] ] )
```

# 1. 모델 1번

```python
dense1 = Dense(units = 100, input_dim = 2, activation = 'relu')
dense2 = Dense(units = 1, input_dim = 100, activation = 'sigmoid')
model = Sequential([dense1, dense2])

model.compile(loss = 'binary_crossentropy', optimizer = Adam(0.01), metrics = ['acc'])

model.fit(x_data, y_data, epochs = 100)
```

# 2. 모델 2번

```python
model = Sequential()
model.add(Dense(units = 100, input_dim = 2, activation = 'relu'))
model.add(Dense(units = 1, input_dim = 100, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = Adam(0.01), metrics = ['acc'])

model.fit(x_data, y_data, epochs = 100)
```

# 3. 모델 3번

```python
input = Input(shape = (2,))
x = Dense(units = 64, activation = 'relu')(input)
x = Dense(units = 32, activation = 'relu')(x)
hx = Dense(units = 1, activation = 'sigmoid')(x)
model = Model(inputs = input, outputs = hx)
model.compile(loss = 'binary_crossentropy', optimizer = Adam(0.01), metrics = ['acc'])
model.summary()
```

```python
model.fit(x_data, y_data, epochs = 100)
```

# 4. 모델 4번 (class 기반)
- 모델 수정하기 간단함, 라이브러리로 활용하기 좋음

```python
class XORModel(Model) : 
    def __init__(self, feature_num = 1, num_class = 1):
        super().__init__(Input(shape = (feature_num,)))
        self.dense1 = Dense(64, activation = 'relu')
        self.dense2 = Dense(32, activation = 'relu')
        self.dense3 = Dense(num_class, activation = 'sigmoid')

    def call(self, xd) :
        print('call')
        x = self.dense1(xd)
        x = self.dense2(x)
        x = self.dense3(x)
        return x
```

```python
model = XORModel(feature_num = 2)
model.compile(loss = 'binary_crossentropy', optimizer = Adam(0.1), metrics = ['acc'])
model.fit(x_data, y_data, epochs = 100)
```
