```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# tensor는 실수와 실수만 계산 가능함 => 실수로 선언하자
x_data = tf.constant([1,2,3], tf.float32)
y_data = tf.constant([1,2,3], tf.float32)
```

```python
x_data
```

```python
def cost(x,y,w):
    hx = w*x # [-1,-1,-1] X [1,2,3] = [-1,-2,-3]
    c = tf.reduce_sum((hx - y)**2) # [-1,-2,-3]-[1,2,3] = [-2, -4, -6] **2 = [4,16,36]
    return c/len(x)
```

```python
print(cost(x_data, y_data, -1))
print(cost(x_data, y_data, 0))
print(cost(x_data, y_data, 1))
print(cost(x_data, y_data, 2))
```

```python
x = tf.constant([1,2,3,4,5], tf.float32)
y = tf.constant([3,5,7,9,11], tf.float32)
```

```python
w = tf.constant(10, tf.float32)
b = tf.constant(10, tf.float32)
epochs = 2000
learning_rate = 0.01
n = len(x)
```

```python
for i in range(epochs):
    hx = w*x + b
    c = tf.reduce_sum((hx-y)**2) / n
    gradientW = tf.reduce_sum((w * x + b - y)*2*x) / n
    gradientB = tf.reduce_sum((w * 1 + b - y)*2) / n
    print('w=',w.numpy(),'b=',b.numpy(), 'w미분',gradientW.numpy(),'b미분:', gradientB.numpy(), 'cost', c.numpy())
    w = w - learning_rate*gradientW
    b = b - learning_rate*gradientB

print('최종w',w.mumpy(),'최종b',b.numpy())
```
