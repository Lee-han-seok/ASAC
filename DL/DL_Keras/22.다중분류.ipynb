```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('data/softmax.txt', sep = ' ', header = None)
df
```

```python
x_data = df.iloc[:,:2]
y_data = df.iloc[:,2:]
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
initX = tf.initializers.GlorotUniform()
w = tf.Variable(initX(shape = [2,3]))
b = tf.Variable(initX(shape = [3]))
```

```python
def compute_loss():
    z = tf.matmul(x, w) + b
    cost_i = tf.nn.softmax_cross_entropy_with_logits(labels = y, logits = z)
    c = tf.reduce_mean(cost_i)
    return c
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss, var_list = [w,b])
    print(i, 'cost:', compute_loss().numpy())
```

```python
def hxFn(xd):
    xd = np.float32(xd)
    z = tf.matmul(xd, w) + b
    hx = tf.nn.softmax(z)
    return hx.numpy()
```
