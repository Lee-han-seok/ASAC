```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
x_data = [1,2,3]
y_data = [1,2,3]
```

```python
plt.scatter(x_data, y_data)

plt.show()
```

```python
# mse
def cost(x,y,w):
    c = 0
    for i in range(len(x)):
        hx = w*x[i]
        c = c + (hx-y[i])**2
    return c/len(x)
```

```python
print(cost(x_data, y_data, -1))
print(cost(x_data, y_data, 0))
print(cost(x_data, y_data, 1))
print(cost(x_data, y_data, 2))
print(cost(x_data, y_data, 3))
```

```python
np.linspace(1,10,3)
```

```python
for w in np.linspace(-3,5,50):
    c = cost(x_data, y_data, w)
    print('w=', w, 'cost:', c)
    plt.plot(w,c,'ro')
plt.xlabel('w')
plt.ylabel('cost(mse)')
plt.show()
```

- w 초기값 아무거나 (랜덤하게)
- w 값의 접선의 기울기를 구한다.(미분)
- w = w - learningrate*미분

미분
- y = 2 미분 0
- y = x 미분 1
- y = 2x 미분 2
- y = x^2 미분 2x
- y = x^n 미분 nx^(n-1)
