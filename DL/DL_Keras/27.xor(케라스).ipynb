```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
import pandas as pd
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0] ] )
```

```python
# w:2x1, b1  z = tf.matmul( x,w) +b )  hx = sigmoid( z )
dense = Dense( units=1, input_dim=2, activation='sigmoid')
model = Sequential( [dense] )
model.compile( loss='binary_crossentropy', optimizer = Adam( 0.1) , metrics=['acc'])
model.fit( x_data, y_data, epochs=1000)
```

```python
# w1:2x100, b1:100  z1 = tf.matmul( x,w1) +b1 )  hx1 = sigmoid( z1 )
dense1 = Dense( units=100, input_dim=2, activation='sigmoid')
# w2:100x1, b1:1  z1 = tf.matmul( x,w2) +b2 )  hx = sigmoid( z1 )
dense2 = Dense( units=1, input_dim=100, activation='sigmoid')
```

```python
hx1 = dense1( x_data ) # z1 = tf.matmul( x_data,w1) +b1 )  hx1 = sigmoid( z1 )
hx = dense2( hx1 ) # z2 = tf.matmul( hx1, w2 ) + b  hx = sigmoid( z2 )
hx
```

```python
x = dense1( x_data )
x = dense2( x )
x
```

```python
dense1 = Dense( units=100, input_dim=2, activation='sigmoid')
dense2 = Dense( units=1, input_dim=100, activation='sigmoid')
model = Sequential([dense1, dense2])
model.compile(loss = 'binary_crossentropy', optimizer = Adam(0.1), metrics = ['acc'])
model.fit(x_data, y_data, epochs = 1000)
```

```python
# 중간층은 활성함수를 relu로 사용
```

```python
def relu(z):
    return 0 if z<0 else z
```

```python
for z in np.linspace(-10, 10, 50):
    r = relu(z)
    print('z-', z, 'relu-', r)
    plt.plot(z, r, 'ro')
plt.show()
```

```python
dense1 = Dense( units=64, input_dim=2, activation='relu')
dense2 = Dense( units=32, activation='relu')
dense3 = Dense( units=1, activation='sigmoid')
model = Sequential([dense1, dense2, dense3])
model.compile(loss = 'binary_crossentropy', optimizer = Adam(0.1), metrics = ['acc'])
model.fit(x_data, y_data, epochs = 200)
```
