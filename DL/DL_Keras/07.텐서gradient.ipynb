```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.optimizers import SGD
```

```python
## 텐서 라이브러리 그대로 : Low Level
## 케라스 라이브러리 이용 : high level
```

```python
x_data = np.array([1,2,3,4,5])
y_data = np.array([3,5,7,9,11])
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w = tf.Variable(10.0)
b = tf.Variable(10.0)
```

```python
def compute_cost():
    hx = w*x+b
    c = tf.reduce_mean((hx-y)**2)
    return c
```

```python
optimizer = SGD(learning_rate = 0.01)
for i in range(2000):
    optimizer.minimize(compute_cost, var_list = [w,b])
    # w.assign_sub(0.01 X 미분값)
    print(i, '코스트:', compute_cost().numpy(), 'w=',w.numpy(), 'b=', b.numpy())
```
