```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer, text_to_word_sequence
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.layers import Dense, SimpleRNN, Embedding, LSTM
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
text = '''하늘에서 오늘 눈이 내리고 있다
그의 눈이 예쁘다
그는 오늘 공부를 한다'''
```

```python
t = Tokenizer()
t.fit_on_texts( [text]) # list of string
```

```python
t.word_index
```

```python
text_to_word_sequence( text )
```

```python
# text_to_word_sequence( text )
s = text.split('\n')
s
```

```python
sentences = []
for line in s:
    print("line:", line)
    encoded = t.texts_to_sequences( [line] )[0]
    print("encoded", encoded)
    for i in range(1, len(encoded) ):
        sentence = encoded[:i+1]
        sentences.append( sentence )
print( sentences )
```

```python
sentences
```

```python
textData = pad_sequences( sentences)
```

```python
textData
```

```python
x = textData[:,:-1]
```

```python
y = textData[:,[-1] ]
```

```python
y.shape
```

```python
x
```

```python
x.reshape(-1,4,1)
```

```python
model = Sequential()
model.add( LSTM(32, activation='relu', input_shape=(4,1)))
model.add( Dense(11, activation='softmax') )
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',
             metrics=['accuracy'])
model.fit( x.reshape(-1,4,1), y , epochs=500)
```
