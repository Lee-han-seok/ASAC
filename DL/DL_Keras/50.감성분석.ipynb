```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU, TimeDistributed
# SimpleRNN -> LSTM (계산이 너무 복잡) -> GRU (계산 간소화, LSTM 정도의 성능) 
from tensorflow.keras import Sequential

import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
```

```python
# 문자를 숫자화 해주는것이 관건
```

```python
x_data = ['좋은 영화','매우 잘만듬', '이건 최악 최악', '너무 못만듬', '굿', '이건 영화 아니다']
y_data = np.array([[1],[1],[0],[0],[1],[0]])
```

```python
cv = CountVectorizer()
result = cv.fit_transform(x_data)
```

```python
print(result)
```

```python
X = result.toarray()
```

```python
cv.vocabulary_
```

```python
X = X.reshape(-1, 9, 1)
```

```python
X.shape
```

```python
model = Sequential()
model.add(LSTM(32, activation = 'relu', input_shape = (9, 1)))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['acc'])

model.summary()
```

```python
model.fit(X, y_data, epochs = 500)
```
