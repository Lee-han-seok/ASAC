```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
x_data = [1,2,3]
y_data = [1,2,3]
```

```python
# mse
def cost(x,y,w):
    c = 0
    for i in range(len(x)):
        hx = w*x[i]
        c = c + (hx-y[i])**2
    return c/len(x)
```

```python
def gradient(x,y,w):
    c = 0
    for i in range(len(x)):
        hx = w*x[i] # 예측값
        c = c + (hx - y[i])*x[i]
    return c/len(x)    
```

```python
w = 10
for epoch in range(100):
    c = cost(x_data, y_data, w)
    g = gradient(x_data, y_data, w)
    print('epoch:',epoch,'w=',w,'기울기:',g,'cost',c)
    w = w - 0.1*g

print('최종w:',w)
```
