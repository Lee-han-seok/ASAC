```python
timesteps = 100
input_features = 32
output_features = 64

inputs = np.random.random((timesteps, input_features)) # 100 X 64 행렬

state_t = np.zeros((output_features,)) # 처음에는 0으로 구성된 값 투입

W = np.random.random((output_features, input_features)) # 64 X 32 행렬
U = np.random.random((output_features, output_features)) # 64 X 64 행렬
b = np.random.random((output_features,)) # 64, 행렬

successive_outputs = [] # (100 X 64) 행렬 
for input_t in inputs:
    output_t = np.tanh(np.dot(W, input_t) + np.dot(U, state_t) + b) # 입력값과 가중치 계산을 통해 도출 된 결괏값
    successive_outputs.append(output_t) # 결괏값을 successive_outputs에 기억해둔다.
    state_t = output_t
    print(state_t) # 64개
final_output_sequence = np.concatenate(successive_outputs, axis = 0)
```
