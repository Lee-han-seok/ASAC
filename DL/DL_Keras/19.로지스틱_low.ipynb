```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
```

```python
x_data = np.array([[1,3], [2,2], [3,1], [4,6], [5,5], [6,4]])
y_data = np.array([[0],[0],[0],[1],[1],[1]])
```

```python
x = tf.constant(x_data,tf.float32)
y = tf.constant(y_data,tf.float32)
```

```python
# 초기화 : xavier, he (최적 초기화 알고리즘)
initX = tf.initializers.GlorotUniform()
w = tf.Variable(initX(shape = [2,1]))
b = tf.Variable(initX(shape = [1]))
```

```python
def compute_loss():
    z = tf.matmul(x, w) + b
    #hx = tf.nn.sigmoid(z)
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits(labels = y, logits = z)
    print('cost_i:', cost_i)
    cost = tf.reduce_mean(cost_i)
    return cost
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss, var_list = [w,b])
    print(i, 'cost:', compute_loss().numpy())
    
```

```python
def hxfn(xd):
    xd = np.float32(xd)
    z = tf.matmul(xd, w) + b
    hx = tf.nn.sigmoid(z)
    return z.numpy(), hx.numpy(), (hx.numpy() > 0.5) + 0
```
