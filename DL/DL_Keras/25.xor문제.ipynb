```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0] ] )
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
initX = tf.initializers.GlorotUniform()
w = tf.Variable( initX( shape=[2,1] ) )
b = tf.Variable( initX( shape=[1] ) )
```

```python
def compute_loss():
    z = tf.matmul( x,w) +b
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits( logits=z, labels=y)
    cost = tf.reduce_mean( cost_i)
    return cost
```

```python
def hxfn( xd ):
    xd = np.float32( xd )
    z = tf.matmul( xd, w) +b
    hx = tf.nn.sigmoid( z )
    return ( hx.numpy()>0.5 ) + 0
```

```python
optimizer = Adam( learning_rate=0.1)
for i in range( 1000):
    optimizer.minimize( compute_loss, var_list=[w,b] )
    pred = hxfn( x )
    acc = (pred == y_data).mean()
    print( i, compute_loss().numpy(), '정확도:', acc )
```
